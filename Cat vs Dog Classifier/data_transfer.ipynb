```python
import os

dest_dir = 'data/train/cats'
os.makedirs(dest_dir, exist_ok=True)
print(f"Ensured directory exists: {dest_dir}")

# Move the first 10,000 images
# Using shell command directly in Python for better control and logging
import subprocess

source_dir = '/content/PetImages/Cat'

# Get the list of files to move
# Using 'find' to get files and 'head' to limit to 10000
command = f'find "{source_dir}" -maxdepth 1 -type f -print0 | head -z -n 10000 | xargs -0 -I {{}} mv {{}} "{dest_dir}"'

try:
    # Execute the command
    result = subprocess.run(command, shell=True, check=True, capture_output=True, text=True)
    print("Files moved successfully.")
    if result.stdout:
        print("STDOUT:", result.stdout)
    if result.stderr:
        print("STDERR:", result.stderr)
except subprocess.CalledProcessError as e:
    print(f"Error moving files: {e}")
    print(f"Command: {e.cmd}")
    print(f"STDOUT: {e.stdout}")
    print(f"STDERR: {e.stderr}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
